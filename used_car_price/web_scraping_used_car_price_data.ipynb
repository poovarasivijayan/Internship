# Ahmedabad:

```python
url1 = "https://www.cardekho.com/used-cars+in+ahmedabad"
driver.get(url1)
driver.maximize_window()
time.sleep(5)
for _ in range(100):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
car_urls= []
url = driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
for i in url:
    car_urls.append(i.get_attribute("href"))
len(car_urls)
#creating the empty lists
name = []
varient = []
engine = []
kilometers = []
car_type= []
owner = []
year = []
price = []

for url in car_urls:
    driver.get(url)
    time.sleep(10)
    
    # fetching name of used cars
    try:
        names = driver.find_elements_by_xpath("//div[@class='gsc_col-xs-12']/h1")
        for j in names:
            name.append(j.text)
        
    except NoSuchElementException:
        name.append('None')
    
    # fetching varient of used cars
    try:
        car_varient = driver.find_elements_by_xpath("//div[@class='variant-name']")
        for car_v in car_varient:
            varient.append(car_v.text)
    except NoSuchElementException:
        varient.append('None')
        
    # fetching engine type driven of used cars
    try:
        engine_type = driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[3]/div/div[2]")
        for engines in engine_type:
            engine.append(engines.text)
    except NoSuchElementException:
        engine.append('None')
        
    # fetching kilometers driven of used cars
    try:
        kilo_d= driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[4]/div/div[2]")
        for kilo in kilo_d:
            kilometers.append(kilo.text)
    except NoSuchElementException:
        kilometers.append('None')
        
    # fetching car_type details of used cars
    try:
        car_t = driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[8]/div/div[2]")
        for types in car_t:
            car_type.append(types.text)
    except NoSuchElementException:
        car_type.append('None')
        
    # fetching owner details of used cars
    try:
        no_of_owner = driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[6]/div/div[2]")
        for owners in no_of_owner:
            owner.append(owners.text)
    except NoSuchElementException:
        owner.append('None')
        
    # fetching year of used cars
    try:
        years = driver.find_elements_by_xpath("//ul[@class='gsc_row detailsList']/li[1]/div/div[2]")
        for loc in years:
            year.append(loc.text)
    except NoSuchElementException:
        year.append('None')
        
    # fetching price of used cars
    try:
        prices = driver.find_elements_by_xpath("//div[@class='priceSection']")
        for pri in prices:
            price.append(pri.text)
    except NoSuchElementException:
        price.append('None')
        
        
#dataframe
df1 = pd.DataFrame({})
df1["name"] = name
df1["varient"] = varient
df1["engine"] = engine
df1["kilometers"] = kilometers
df1["car_type"] = car_type
df1["location"] = location
df1["year"]=year
df1["price"] = price
```
