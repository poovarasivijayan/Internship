```python
click_button = driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td/span/b/a")
click_button.click()
```

```python
dataset_name = []
data_type = []
task = []
attribute = []
no_of_instance = []
no_of_attributes = []
year = []

try:
    dataset = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
    for i in dataset:
        dataset_name.append(i.text)
except NoSuchElementException:
    dataset_name.append("---")
```

```python
try:
    data = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
    for i in data[1:]:
        data_type.append(i.text)
except NoSuchElementException:
    data_type.append("---")
    
try:
    tasks = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
    for i in tasks[1:]:
        task.append(i.text)
except NoSuchElementException:
    task.append("---")
    
try:
    attri = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
    for i in attri[1:]:
        attribute.append(i.text)
except NoSuchElementException:
    attribute.append("---")
    
try:
    instance = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
    for i in instance[1:]:
        no_of_instance.append(i.text)
except NoSuchElementException:
    no_of_instance.append("---")
    
try:
    attribu = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
    for i in attribu[1:]:
        no_of_attributes.append(i.text)
except NoSuchElementException:
    no_of_attributes.append("---")
    
try:
    years = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]")
    for i in years[1:]:
        year.append(i.text)
except NoSuchElementException:
    year.append("---")
    
df = pd.DataFrame({"Dataset Name":dataset_name,
                  "Data Type":data_type,
                  "Task":task,
                   "No of Instance":no_of_instance,
                  "Attribute":attribute,
                  "No of Attributes":no_of_attributes,
                  "Year":year})
df
```
